```python
data1 = pd.read_csv ('kddcup.csv', header=None)

from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
data1[1] = class_le.fit_transform(data1[1].values)
data1[2] = class_le.fit_transform(data1[2].values)
data1[3] = class_le.fit_transform(data1[3].values)
data1[41] = class_le.fit_transform(data1[41].values)

data1
```

```python
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
XX = data1.iloc[:,0:41]
X_train_std = stdsc.fit_transform(data1.iloc[:,0:41].values)
```

```python
from sklearn.preprocessing import LabelEncoder
y = data1.iloc[:,41]
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
X_train_lda = lda.fit_transform(X_train_std, y)
X_train_lda.shape
```

```python
data2=pd.DataFrame(X_train_lda)
data2['class']=y

data2.rename(columns={0: "LD1", 1: "LD2"}, inplace=True)
```

```python
sns.lmplot(x="LD1", y="LD2", data=data2, hue='class',fit_reg=False,legend=True)
plt.show()
```

```python
from sklearn.decomposition import PCA 
pca = PCA().fit(X_train_std)
temp = np.cumsum(pca.explained_variance_ratio_)
a = 0
for i in temp:
    a=a+1
    if i>0.95:
        b=a
        break
pca = PCA(n_components=b)
X_train_pca = pca.fit_transform(X_train_std, y)
X_train_pca.shape
```

```python
data3=pd.DataFrame(X_train_pca)
data3['class']=y
data3.rename(columns={0: "PC1", 1: "PC2"}, inplace=True)
```

```python
sns.lmplot(x="PC1", y="PC2", data=data3, hue='class',fit_reg=False,legend=True)
plt.show()
```

```python
from sklearn.decomposition import NMF

nmf = NMF(n_components=22)
X_train_nmf = nmf.fit_transform(XX, y)
X_train_nmf.shape
```

```python
data4=pd.DataFrame(X_train_nmf)
data4['class']=y
data4.rename(columns={0: "NMF1", 1: "NMF2"}, inplace=True)
```

```python
sns.lmplot(x="NMF1", y="NMF2", data=data4, hue='class',fit_reg=False,legend=True)
plt.show()
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
sfs = SequentialFeatureSelector(knn, n_features_to_select=22, direction='backward')
X_train_sfs = sfs.fit_transform(X_train_std, y)
X_train_sfs.shape
```

```python
data5=pd.DataFrame(X_train_sfs)
data5['class']=y
data5.rename(columns={0: "SFS1", 1: "SFS2"}, inplace=True)
```

```python
sns.lmplot(x="SFS1", y="SFS2", data=data5, hue='class',fit_reg=False,legend=True)
plt.show()
```

Perform different classifier techniques on Original data, LDA data, PCA data and NMF data.
Comparison will show LDA is better

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import log_loss
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_std, y, test_size=0.20, random_state=42)

model_orig = LogisticRegression()
model_orig.fit(X_train,y_train)

yhat_prob_1=model_orig.predict_proba(X_test)
yhat_1=model_orig.predict(X_test)
print (classification_report(y_test, yhat_1))
confusion_matrix(y_test, yhat_1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_lda, y, test_size=0.20, random_state=42)

model_LDA = LogisticRegression()
model_LDA.fit(X_train,y_train)

yhat_prob_1=model_LDA.predict_proba(X_test)
yhat_1=model_LDA.predict(X_test)
print (classification_report(y_test, yhat_1))
confusion_matrix(y_test, yhat_1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_pca, y, test_size=0.20, random_state=42)

model_PCA = LogisticRegression()
model_PCA.fit(X_train,y_train)

yhat_prob_1=model_PCA.predict_proba(X_test)
yhat_1=model_PCA.predict(X_test)
print (classification_report(y_test, yhat_1))
confusion_matrix(y_test, yhat_1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_nmf, y, test_size=0.20, random_state=42)

model_NMF = LogisticRegression()
model_NMF.fit(X_train,y_train)

yhat_prob_1=model_NMF.predict_proba(X_test)
yhat_1=model_NMF.predict(X_test)
print (classification_report(y_test, yhat_1))
confusion_matrix(y_test, yhat_1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_sfs, y, test_size=0.20, random_state=42)

model_SFS = LogisticRegression()
model_SFS.fit(X_train,y_train)

yhat_prob_1=model_SFS.predict_proba(X_test)
yhat_1=model_SFS.predict(X_test)
print (classification_report(y_test, yhat_1))
confusion_matrix(y_test, yhat_1)
```
